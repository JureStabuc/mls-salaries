```python
import pandas as pd
from pandas import DataFrame
```

```python
for year in range(2007, 2018):
    data = pd.read_csv("mls-salaries-{0}.csv".format(year))
    df = DataFrame(data, columns=["last_name", "first_name", "guaranteed_compensation"])
    df = df.nlargest(10, "guaranteed_compensation", keep='first')
    df["guaranteed_compensation"] = df["guaranteed_compensation"].astype(int)
    print(df)
    df.to_csv("top10/mls-top10-{0}.csv".format(year), index = False)
```
